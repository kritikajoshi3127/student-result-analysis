```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df= pd.read_csv("archive.zip")
print (df.head())
```

```python
sns.countplot(data = df, x="Gender")
plt.show()
```

```python
plt.figure(figsize=(5,5))
ax =sns.countplot(data = df, x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

# from above chart we have number of female chart is more than number of males


```python
gb = df.groupby("Height(CM)").agg({"10th Mark":"mean","12th Mark":"mean","college mark":"mean"})
print(gb)
```

```python
plt.figure(figsize=(10,10))
plt.title("relation between student marks and height")
sns.heatmap(gb,annot=True)
plt.show
```

# IN CLASS 10TH WE SEE LARGER THE HEIGHT MORE THE SCORE BUT IN 12TH AND COLLEGEWE SEES ALL SCORE ARE ALMOST SIMILAR FOR ALL HEIGHTS 


```python
gb = df.groupby("Certification Course").agg({"10th Mark":"mean","12th Mark":"mean","college mark":"mean"})
print(gb)
```

```python
plt.figure(figsize=(10,10))
plt.title("relation between student marks and they have done any course")
sns.heatmap(gb,annot=True)
plt.show
```

# STUDENT WHO HAVE DONE  CERTIFIED COURSE HAVE MORE MARKS THAN THE STUDENT WHO DONOT HAVE DONE THE COURSE

```python
gb = df.groupby("daily studing time").agg({"10th Mark":"mean","12th Mark":"mean","college mark":"mean"})
print(gb)
```

```python
plt.figure(figsize=(10,10))
plt.title("relation between student marks and daily studing time ")
sns.heatmap(gb,annot=True)
plt.show
```
